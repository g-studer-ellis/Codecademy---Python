### What is the relationships between medical costs and other columns?

#### Investigation

What's the relationship between charges and patient characteristics (age, sex, bmi, children, and region)?  

First I will use Spearman's correlation to determine if charges are correlated with any of the above characteristics. I will determine positive, negative, or neutral correlation. I will also investigate the p-value of the correlations.

I will visualize the relationships between charges and characteristics using scatter plots and, where appropriate, violin plots..

I could attempt to plot all of the relationships on one plot, however this would be too cluttered.  
Instead I will create blocks of code which focus on one patient characteristic and plot its relationship to the other characteristics individually

##### Spearman's Correlation

```python
age_corr = sts.spearmanr(insurance_new['charges'], insurance_new['age'])
print('Charges and Age: ' + str(age_corr))

sex_corr = sts.spearmanr(insurance_new['charges'], insurance_new['sex'])
print('Charges and Sex: ' + str(sex_corr))

bmi_corr = sts.spearmanr(insurance_new['charges'], insurance_new['bmi'])
print('Charges and BMI: ' + str(bmi_corr))

children_corr = sts.spearmanr(insurance_new['charges'], insurance_new['children'])
print('Charges and Number of Cildren: ' + str(children_corr))

smoker_corr = sts.spearmanr(insurance_new['charges'], insurance_new['smoker'])
print('Charges and Smoker Status: ' + str(smoker_corr))

region_corr = sts.spearmanr(insurance_new['charges'], insurance_new['region'])
print('Charges and Region: ' + str(region_corr))
```

The correlations between Charges and patient Characteristics are:
- Age: moderate, statistically significant
- Sex: very weak, not statistically significant
- BMI: very weak, statistically significant
- Number of Children: very weak, statistically significant
- Smoker Status: strong, statistically significant
- Region: very weak, not statistically significant

##### Visualizing Relationships

```python
# Letting python know I'm going to use seaborn plots, and what schematic options I want to use
sns.set()
```

```python
def make_scatter_plots(df, char):
    char_list = ['age', 'sex', 'bmi', 'children', 'smoker', 'region']
    sns.scatterplot(data=df, x=df[char], y=df['charges'])
    plt.show()
    plt.savefig(str(char)+'_charges_scatter.png')
    plt.clf()
    for i in char_list:
        if i != char:
            sns.scatterplot(data=df, x=char, y=df['charges'], hue=df[i])
            plt.show()
            plt.savefig(str(char)+'_charges_'+str(i)+'_scatter.png')
            plt.clf()
        else:
            print('Cannot hue the x-axis')
```

### Do smokers have higher medical costs than non-smokers?

#### Quantifying the costs for smokers and non-smokers

- What are the summary stats for charges for non-smokers?  
- What are the summary stats for charges for a smoker?  

```python
smokers = insurance_new.loc[insurance_new['smoker'] == 'yes']
smokers_charges = insurance_new.loc[insurance_new['smoker'] == 'yes', 'charges']

non_smokers = insurance_new.loc[insurance_new['smoker'] == 'no']
non_smokers_charges = insurance_new.loc[insurance_new['smoker'] == 'no', 'charges']
```

```python
smokers_charges_stats = smokers_charges.describe(include='all')
non_smokers_charges_stats = non_smokers_charges.describe(include='all')
```

```python
charges_comparison = pd.DataFrame()
charges_comparison['smokers'] = smokers_charges_stats
charges_comparison['non_smokers'] = non_smokers_charges_stats
charges_comparison
```

#### Comparing the costs for smokers and non-smokers

```python
tstat, pval = sts.ttest_ind(smokers_charges, non_smokers_charges, equal_var = False)
one_sided_pval = pval/2
print(one_sided_pval)
```

#### Inferences

The p-value from the Welch's t-test rejects the null hypothesis that the smokers mean charges is not greater than the non-smokers mean charges.  

In more plain test: smokers do have higher charges than non-smokers.

### Do smokers have other indicators for higher charges?

From section 4.3.1.1 I determined that increased age and increased BMI can have a correlation with increased insurance charges. For this section I will determine if smokers have increased age and or increased BMI.

#### Smokers vs. non-smokers and BMI

```python
smokers_bmi = insurance_new.loc[insurance_new['smoker'] == 'yes', 'bmi']
non_smokers_bmi = insurance_new.loc[insurance_new['smoker'] == 'no', 'bmi']

smokers_bmi_stats = smokers_bmi.describe(include='all')
non_smokers_bmi_stats = non_smokers_bmi.describe(include='all')

bmi_comparison = pd.DataFrame()
bmi_comparison['smokers'] = smokers_bmi_stats
bmi_comparison['non_smokers'] = non_smokers_bmi_stats

tstat, pval = sts.ttest_ind(smokers_bmi, non_smokers_bmi, equal_var = False)
one_sided_pval = pval/2
print(one_sided_pval)
```

The p-value from the Welch's t-test fails to reject the null hypothesis that the smokers mean BMI is not greater than the non-smokers mean BMI.  

In more plain test: smokers do not have higher BMIs than non-smokers.

#### Smokers vs. non-smokers and age

```python
smokers_age = insurance_new.loc[insurance_new['smoker'] == 'yes', 'age']
non_smokers_age = insurance_new.loc[insurance_new['smoker'] == 'no', 'age']

smokers_age_stats = smokers_age.describe(include='all')
non_smokers_age_stats = non_smokers_age.describe(include='all')

age_comparison = pd.DataFrame()
age_comparison['smokers'] = smokers_age_stats
age_comparison['non_smokers'] = non_smokers_age_stats

tstat, pval = sts.ttest_ind(smokers_age, non_smokers_age, equal_var = False)
one_sided_pval = pval/2
print(one_sided_pval)
```

The p-value from the Welch's t-test fails to reject the null hypothesis that the smokers mean age is not greater than the non-smokers mean age.  

In more plain test: smokers are not older than non-smokers.
